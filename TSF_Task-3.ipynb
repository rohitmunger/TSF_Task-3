#### Importing required libraries

```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.tree import plot_tree
import pandas as pd
import matplotlib.pyplot as plt
```

#### Loading the Iris Dataset

```python
iris_data = datasets.load_iris()
```

```python
df = pd.DataFrame(iris_data.data, columns= iris_data.feature_names)
df.head(5)
```

```python
# Getting basic insights of the data
df.info()
```

```python
# Statisstical info
df.describe().T
```

#### Visualizing the dataset

```python
pair_plot = pd.plotting.scatter_matrix(df,marker='o',figsize=(12,10),color='red')
```

#### Data Preparation for the Dicision Tree Classifier

```python
X = df
y = iris_data.target
```

```python
print(y) # These are the target variables
```

#### Creating the Decision Tree Classifier

```python
desc_tree = DecisionTreeClassifier()
```

```python
# Training the classifier with X, y
desc_tree.fit(X, y)
```

```python
# Evaluating the performance of the classifier
accuracy = cross_val_score(desc_tree, X, y,cv=3).mean()
```

```python
print(f"\033[1m Accuracy of The Classifier is: {round(accuracy*100,2)} % \033[1m")
```

#### Visualizing the Decision Tree

```python
# Setting the figure size
plt.figure(figsize=(20,10))
tree = plot_tree(desc_tree, filled=True)
```
